#### В каждой строке данных содержится следующая информация

### Подготовка и проверка данных

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

columns = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]
data = pd.read_csv("data/housing.xls", header=None, delimiter=r"\s+", names=columns)

print("Проверка лишних данных")
data.info()

df = pd.DataFrame(data)
correlation_matrix = df.corr()
print("Матрица корреляции:\n", correlation_matrix)


print("Выбросы:\n", data['MEDV'].describe())

X_train, X_test, y_train, y_test = train_test_split(data.drop("MEDV", axis=1), data["MEDV"], test_size=0.2)
```

### Обучение моделей

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(X_train, y_train)

from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor()
forest_reg.fit(X_train, y_train)
```

### Оценка моделей по метрикам

```python
lin_pred = lin_reg.predict(X_test)
tree_pred = tree_reg.predict(X_test)
forest_pred = forest_reg.predict(X_test)
models = [lin_pred, tree_pred, forest_pred]

print(f"{'Model':<8} {'Linear':<18} {'Tree':<18} {'Forest':<0}")
print(f"MSE:\t", *[np.mean((y_test - model)**2) for model in models])
print(f"RMSE:\t", *[np.sqrt(np.mean((y_test - model)**2)) for model in models])
print(f"MAE:\t", *[np.mean(np.abs(y_test - model)) for model in models])
```

```python
importance_lin = pd.DataFrame({
    "Feature": data.drop("MEDV", axis=1).columns,
    "Coefficient": lin_reg.coef_
}).sort_values(by="Coefficient", key=abs, ascending=False)

importance_tree = pd.DataFrame({
    "Feature": data.drop("MEDV", axis=1).columns,
    "Coefficient": tree_reg.feature_importances_
}).sort_values(by="Coefficient", ascending=False)

importance_forest = pd.DataFrame({
    "Feature": data.drop("MEDV", axis=1).columns,
    "Coefficient": forest_reg.feature_importances_
}).sort_values(by="Coefficient", ascending=False)

print(f"Linear coefficients:\n{importance_lin}\nTree coefficients:\n{importance_tree}\nForest coefficients:\n{importance_forest}")
```

### Выводы

Лучшая модель: случайный лес (наименьшие ошибки по всем метрикам)\
Важность признаков в модели\
Линейная: NOX. Чем выше NOX, тем дешевле квартиры. B. Почти не влияет\
Дерево: RM. RM самый важный параметр. ZN почти не влияет\
Лес: RM и LSTAT. Почти одинаково важны
